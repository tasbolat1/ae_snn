```python
import pickle
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
from os import listdir
from os.path import isdir, join
import matplotlib.pyplot as plt
```

```python
mypath = '../ae_snn_results/'
```

```python
run_dir = [mypath+f for f in listdir(mypath) if isdir(join(mypath, f))]
```

```python
class RUN():
    def __init__(self, run_path):
        self.run_name = run_path[-32:]
        self.run_path = run_path
        event_acc = EventAccumulator(self.run_path)
        event_acc.Reload()
        self.w_times_train, self.x_train, self.vals_train = zip(*event_acc.Scalars('loss/train'))
        self.w_times_test, self.x_test, self.vals_test = zip(*event_acc.Scalars('loss/test'))
        self.epochs = len(self.x_train)
        
        self.args = pickle.load(open(self.run_path+'/args.pkl', 'rb'))
        
    def plot(self, _ax, color='b'):
        _ax.plot(self.x_train, self.vals_train, color)
        _ax.plot(self.x_test, self.vals_test, color+'--')
```

```python
run1 = RUN(run_dir[0])
```

```python
run1.args.lr
```

```python
fig, ax = plt.subplots(6, figsize=(15,30))
for each_run in run_dir:
    run = RUN(each_run)
    if run.args.lr == 0.001 and run.args.model_type == 1 and run.args.hidden_size==32:
        run.plot(ax[0])
    elif run.args.lr == 0.01 and run.args.model_type == 1 and run.args.hidden_size==32:
        run.plot(ax[1])
    if run.args.lr == 0.001 and run.args.model_type == 1 and run.args.hidden_size==64:
        run.plot(ax[2])
    elif run.args.lr == 0.01 and run.args.model_type == 1 and run.args.hidden_size==64:
        run.plot(ax[3])
    if run.args.lr == 0.001 and run.args.model_type == 1 and run.args.hidden_size==20:
        run.plot(ax[4])
    elif run.args.lr == 0.01 and run.args.model_type == 1 and run.args.hidden_size==20:
        run.plot(ax[5])
    ax[0].set_title('type=1 ,lr=0.001, hidden_size=32')
    ax[1].set_title('type=1 ,lr=0.01, hidden_size=32')
    ax[2].set_title('type=1 ,lr=0.001, hidden_size=64')
    ax[3].set_title('type=1 ,lr=0.01, hidden_size=64')
    ax[4].set_title('type=1 ,lr=0.001, hidden_size=20')
    ax[5].set_title('type=1 ,lr=0.01, hidden_size=20')
```

```python
fig, ax = plt.subplots(4, figsize=(15,20))
for each_run in run_dir:
    run = RUN(each_run)
    if run.args.lr == 0.001 and run.args.model_type == 2 and run.args.hidden_size==32:
        run.plot(ax[0])
    elif run.args.lr == 0.01 and run.args.model_type == 2 and run.args.hidden_size==32:
        run.plot(ax[1])
    if run.args.lr == 0.001 and run.args.model_type == 2 and run.args.hidden_size==20:
        run.plot(ax[2])
    elif run.args.lr == 0.01 and run.args.model_type == 2 and run.args.hidden_size==20:
        run.plot(ax[3])
    ax[0].set_title('type=2 ,lr=0.001, hidden_size=32')
    ax[1].set_title('type=2 ,lr=0.01, hidden_size=32')
    ax[2].set_title('type=2 ,lr=0.001, hidden_size=20')
    ax[3].set_title('type=2 ,lr=0.01, hidden_size=20')
```
